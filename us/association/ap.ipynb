```python
import pandas as pd
import numpy as np
```

```python


# List of possible grocery items
grocery_items = [
    'apple', 'banana', 'bread', 'milk', 'cheese', 'eggs', 'flour', 'sugar', 'salt', 'pepper',
    'rice', 'pasta', 'chicken', 'beef', 'pork', 'fish', 'tomato', 'lettuce', 'potato', 'onion',
    'carrot', 'broccoli', 'yogurt', 'butter', 'jam', 'peanut butter', 'honey', 'tea', 'coffee', 'juice'
]

# Number of rows in the dataset
num_rows = 100

# Generating random grocery items for each column
data = {f'data{i+1}': np.random.choice(grocery_items, num_rows) for i in range(10)}

# Creating a DataFrame
df = pd.DataFrame(data)

# Randomly setting some fields to empty values
nan_indices = np.random.choice([True, False], size=(num_rows, 10), p=[0.1, 0.9])
dataset = df.mask(nan_indices)
```

```python
market = []
for i in range(0,dataset.shape[0]):
    customer = []
    for j in dataset.columns:
        if type(dataset[j][i]) == str:
            customer.append(dataset[j][i])
    market.append(customer)
```

```python
# market
```

```python
l = []
for i in market:
    for j in i:
        # print(j)
        l.append(j)
```
