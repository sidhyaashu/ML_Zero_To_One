Dendrogram & Euclidean distance

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"D:\ML\WsCubeTech\us\clustering\data\iris_csv.csv")
dataset.head(5)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dataset,method="single",metric="euclidean"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage="single")
dataset["pre"]=ac.fit_predict(dataset)
```

```python
dataset.head(3)
```

```python
sns.pairplot(data=dataset,hue="pre")
plt.show()
```
