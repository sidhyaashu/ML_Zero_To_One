```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Generate random data
cgpa = np.random.uniform(6.0, 10.0, 1000)  # CGPA between 6.0 and 10.0
score = np.random.randint(60, 101, 1000)   # Scores between 60 and 100
placed = np.random.choice([1, 0], 1000)  # Random boolean values

# Create DataFrame
data = {
    'cgpa': cgpa,
    'score': score,
    'placed': placed
}

dataset = pd.DataFrame(data)

# Display first 5 rows of the DataFrame
print(dataset.head(5))
```

```python
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="placed")
plt.legend(loc=1)
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=3)
```

```python
pf.fit(x)
```

```python
x = pd.DataFrame(pf.transform(x))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
