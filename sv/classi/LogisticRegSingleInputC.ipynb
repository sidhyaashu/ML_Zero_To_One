```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Set a seed for reproducibility
np.random.seed(42)

# Generate sample data
data = {
    'age': np.random.randint(18, 70, size=120),
    'salary': np.random.randint(30000, 150000, size=120),
    'purchased': np.random.randint(0, 2, size=120)
}
dataset = pd.DataFrame(data)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="age",y="purchased",data=dataset)
plt.show()
```

```python
x = dataset[["age"]]
y = dataset["purchased"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
sns.scatterplot(x="age",y="purchased",data=dataset)
sns.lineplot(x="age",y=lr.predict(x),data=dataset,color="r")
plt.show()
```
