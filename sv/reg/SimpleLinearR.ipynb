```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv(r"D:\ML\WsCubeTech\data\placement\placement.csv")
dataset.head(3)
```

```python
dataset.isnull().sum().sum()
```

```python
sns.scatterplot(x="Hours",y="Scores",data=dataset)
plt.show()
```

```python
x = dataset[["Hours"]]
y = dataset[["Scores"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100 #Accuracy
```

```python
lr.predict([[2.5]])
```

```python
y_prd = lr.predict(x)
```

```python
sns.scatterplot(x="Hours",y="Scores",data=dataset)
plt.plot(dataset["Hours"],y_prd,c="red")
plt.legend(["Original data","Predict Line"])
plt.show()
```
