```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
#Naive - Features
#Bayes - Depend on probability bayes theorem
```

```python
# Types of Naive Bayes model
# 1. Gaussian - data in normal distribution
# 2. Multinomial - features are in binary variable
# 3. bernoulli - text data / discrete data
```

```python
dataset = pd.read_csv(r"D:\ML\WsCubeTech\data\plc\plc.csv")
# dataset.head(5)
```

```python
# Generate random data
cgpa = np.random.uniform(6.0, 10.0, 100)  # CGPA between 6.0 and 10.0
score = np.random.randint(60, 101, 100)   # Scores between 60 and 100
placed = np.random.choice([1, 0], 100)  # Random boolean values

# Create DataFrame
data = {
    'cgpa': cgpa,
    'score': score,
    'placed': placed
}
```

```python
dataset = pd.DataFrame(data)
```

```python
dataset.head(5)
```

```python
sns.kdeplot(data=dataset["score"]) #to check data is in normal distribution
plt.show()
```

```python
sns.kdeplot(data=dataset["cgpa"]) #to check data is in normal distribution
plt.show()
```

```python
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
mnb.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
bnb.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
