```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import numpy as np

# Generating sample data
np.random.seed(42)  # For reproducibility

# CGPA ranges from 6.0 to 10.0
cgpa = np.round(np.random.uniform(6.0, 10.0, 100), 2)

# Package ranges from 5.0 to 15.0, increasing with CGPA
package = 5.0 + (cgpa - 6.0) * (15.0 - 5.0) / (10.0 - 6.0)
# Adding some noise to the package
package += np.random.normal(0, 0.5, 100)
package = np.round(package, 2)

# Creating a DataFrame
data = {
    'cgpa': cgpa,
    'pac': package
}

dataset = pd.DataFrame(data)
```

```python
sns.scatterplot(x="cgpa",y="pac",data=dataset)
plt.show()
```

```python
x = dataset[["cgpa"]]
y = dataset["pac"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.svm import SVR
```

```python
sv = SVR(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sns.scatterplot(x="cgpa",y="pac",data=dataset)
plt.plot(dataset["cgpa"],sv.predict(x),color ="red" )
plt.show()
```
