```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import numpy as np

# Generating sample data
np.random.seed(42)  # For reproducibility

cgpa = np.round(np.random.uniform(0, 10, 100), 2)
score = np.round(np.random.uniform(0, 10, 100), 2)
placed = np.random.randint(0, 2, 100)

# Creating a DataFrame
data = {
    'CGPA': cgpa,
    'Score': score,
    'Placed': placed
}

dataset = pd.DataFrame(data)
```

```python
sns.scatterplot(x="CGPA",y="Score",hue="Placed",data=dataset)
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Placed"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
# sv = SVC(kernel="linear")
sv = SVC(kernel="rbf")
sv.fit(x_train,y_train)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

```python
sv.score(x_test,y_test)*100
```
