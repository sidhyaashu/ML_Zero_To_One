```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
```

```python
cgpa = np.random.randint(18, 70, 100)  # CGPA between 6.0 and 10.0
score = np.random.randint(10000, 100000, 100)   # Scores between 60 and 100
placed = np.random.choice([1, 0], 100)  # Random boolean values

# Create DataFrame
data = {
    'age': cgpa,
    'salary': score,
    'purches': placed
}
```

```python
dataset = pd.DataFrame(data)
dataset.head(5)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["purches"]
```

```python
sns.scatterplot(x="age",y="salary",hue="purches",data=dataset)
plt.show() #it is not idol dataset for this algorithms
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn)
plt.show()
```
