```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
cgpa = np.random.randint(18, 70, 100)  # CGPA between 6.0 and 10.0
score = np.random.randint(10000, 100000, 100)   # Scores between 60 and 100
placed = np.random.choice([1, 0], 100)  # Random boolean values

# Create DataFrame
data = {
    'age': cgpa,
    'salary': score,
    'purches': placed
}
```

```python
dataset = pd.DataFrame(data)
dataset.head(5)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["purches"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=dt)
plt.show()
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.predict([[20,32000]])
```

```python
from sklearn.tree import plot_tree
```

```python
# plt.figure(figsize=(20,20))
plot_tree(dt)
# plt.savefig("tree.jpg")
plt.show()
```

```python
# Pre - Puning
dt = DecisionTreeClassifier(max_depth=5) # max_dept value is user define
```

```python
#Post - Puning
for i in range(1,20):
    dt2 = DecisionTreeClassifier(max_depth=i)
    dt2.fit(x_train,y_train)
    print(dt2.score(x_train,y_train)*100,dt2.score(x_test,y_test)*100,i)
```
